### Analysis

- The trial tested 10 drugs. All the drugs were tested on 25 mice apart from Stelasyn which was tested on 24 mice in the study. The data for one mouse (ID G989) was duplicated and so was removed from the analysis. In the remaining data there were slightly more male mice (51%) then female mice (49%).

- The data shows that the drugs Capomulin and Ramicane were the most promising of the those tested. The tumours all started at the same volume but at every step of the process the mice that were give Capomulin and Ramicane showed smaller tumour volumes as shown by the mean and median values of the tumour volumes over the whole process. The standard deviation and standard error were the smallest for these two drugs, showing that the treatments worked for all the mice that were given these drugs. This is further confirmed by the fact that there were no outliers in the final tumour sizes of these two drugs, so all the mice responded to the treatment.

- The box plots show that out of the two most promising drugs, Ramicane performed marginally better. The median final tumour volume was lower than that of Capomulin and although the interquartile range was slightly larger, the negative whisker was longer than the positive one, indicating a skew towards smaller final tumour volumes.

- The mice tested with Infubinol responded less well, their final median tumour volume was much higher than that of Capomulin and Ramicane. However there was one outlier which showed a significantly lower tumour volume after treatment, corresponding to the median value of Ramicane. This shows that in some cases Infubinol may be an appropriate drug to use.

- The data also showed that there was a significant correlation between average tumour volume and the weight of the mouse. The heavier the mouse the larger the average tumour volume was. This may affect treatment plans.

- Therefore in summary, Pymaceuticals drug of interest, Capomulin, performed well against all the other drugs except Ramicane. Even then the difference between Capomulin and Ramicane was marginal. The varaible that was not explored was judging the performance of Capomulin and Ramicane with regards to mouse weight. Weight seems to have a significant effect on tumour size, therefore it will be interesting to what effect weight had on the performance of Capomulin and Ramicane. 


```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import random

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame
study_data = pd.merge(study_results,mouse_metadata,how="left",on=["Mouse ID"])
study_data.head()
```

```python
# Checking the number of mice.
num_mice = study_data["Mouse ID"].nunique()
num_mice
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
duplicated_value = study_data[study_data.duplicated(subset=["Mouse ID","Timepoint"],keep=False)==True]
duplicated_value["Mouse ID"].unique()
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
duplicated_value
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
study_data_clean = study_data.drop(study_data[study_data['Mouse ID'] == duplicated_value.iloc[0,0]].index)
study_data_clean.head()
```

```python
# Checking the number of mice in the clean DataFrame.
num_mice = study_data_clean["Mouse ID"].nunique()
num_mice
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary DataFrame.

study_data_groupby = study_data_clean.groupby("Drug Regimen")
mean_tumour_vol = study_data_groupby["Tumor Volume (mm3)"].mean()
median_tumour_vol = study_data_groupby["Tumor Volume (mm3)"].median()
variance_tumour_vol = study_data_groupby["Tumor Volume (mm3)"].var()
std_tumour_vol = study_data_groupby["Tumor Volume (mm3)"].std()
sem_tumour_vol = study_data_groupby["Tumor Volume (mm3)"].sem()

tumour_data = pd.DataFrame({"Mean Tumour Volume":mean_tumour_vol,
                            "Median Tumour Volume": median_tumour_vol,
                            "Tumour Volume Variance":variance_tumour_vol,
                            "Tumour Volume St. Dev.":std_tumour_vol,
                            "Tumour Volume Std. Err.":sem_tumour_vol})
tumour_data
```

```python
# A more advanced method to generate a summary statistics table of mean, median, variance, standard deviation,
# and SEM of the tumor volume for each regimen (only one method is required in the solution)

# Using the aggregation method, produce the same summary statistics in a single line

tumour_data = study_data_clean.groupby("Drug Regimen")["Tumor Volume (mm3)"].agg(['mean','median','var','std','sem'])
tumour_data.columns = pd.MultiIndex.from_product([["Tumor Volume (mm3)"],tumour_data.columns])
tumour_data
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse
max_timepoint = study_data_clean.groupby("Mouse ID").max("Timepoint")["Timepoint"]

# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
final_tumour_data = pd.merge(max_timepoint,study_data_clean,on=["Mouse ID","Timepoint"],how="inner")
final_tumour_data
```

```python
# Put treatments into a list for for loop (and later for plot labels)
treatments = ["Capomulin","Ramicane","Infubinol","Ceftamin"]

# Create empty list to fill with tumor vol data (for plotting)
tumour_vol_data_Capomulin = []
tumour_vol_data_Ramicane = []
tumour_vol_data_Infubinol = []
tumour_vol_data_Ceftamin = []
outliers_Capomulin = []
outliers_Ramicane = []
outliers_Infubinol = []
outliers_Ceftamin = []

# Calculate the IQR and quantitatively determine if there are any potential outliers. 
for mouse in range(0,len(final_tumour_data)-1): 
    # Locate the rows which contain mice on each drug and get the tumor volumes
    # add subset      
    if final_tumour_data.loc[mouse]["Drug Regimen"]=="Capomulin":
        tumour_vol_data_Capomulin.append(final_tumour_data.loc[mouse]["Tumor Volume (mm3)"])
    elif final_tumour_data.loc[mouse]["Drug Regimen"]=="Ramicane":
        tumour_vol_data_Ramicane.append(final_tumour_data.loc[mouse]["Tumor Volume (mm3)"])
    elif final_tumour_data.loc[mouse]["Drug Regimen"]=="Infubinol":
        tumour_vol_data_Infubinol.append(final_tumour_data.loc[mouse]["Tumor Volume (mm3)"])
    elif final_tumour_data.loc[mouse]["Drug Regimen"]=="Ceftamin":
        tumour_vol_data_Ceftamin.append(final_tumour_data.loc[mouse]["Tumor Volume (mm3)"])        
    
# Determine outliers using upper and lower bounds
final_tumour_data_by_treatment = pd.DataFrame({ "Capomulin":tumour_vol_data_Capomulin,
                                                "Ramicane":tumour_vol_data_Ramicane,
                                                "Infubinol":tumour_vol_data_Infubinol,
                                                "Ceftamin":tumour_vol_data_Ceftamin})


final_tumour_data_by_treatment_quartile = final_tumour_data_by_treatment.quantile([0.25,0.5,0.75])
lowerq = final_tumour_data_by_treatment_quartile.iloc[0,:]
upperq = final_tumour_data_by_treatment_quartile.iloc[2,:]
iqr = upperq - lowerq
lower_bound = lowerq - (1.5*iqr)
upper_bound = upperq + (1.5*iqr)

[outliers_Capomulin.append(outlier) for outlier in tumour_vol_data_Capomulin if (outlier<lower_bound["Capomulin"] or outlier >upper_bound["Capomulin"])]
[outliers_Ramicane.append(outlier) for outlier in tumour_vol_data_Ramicane if (outlier<lower_bound["Ramicane"] or outlier >upper_bound["Ramicane"])]
[outliers_Infubinol.append(outlier) for outlier in tumour_vol_data_Infubinol if (outlier<lower_bound["Infubinol"] or outlier >upper_bound["Infubinol"])]
[outliers_Ceftamin.append(outlier) for outlier in tumour_vol_data_Ceftamin if (outlier<lower_bound["Ceftamin"] or outlier >upper_bound["Ceftamin"])]

print(f"Capomulin's potential outliers: {outliers_Capomulin}")
print(f"Ramicane's potential outliers: {outliers_Ramicane}")
print(f"Infubinol's potential outliers: {outliers_Infubinol}")
print(f"Ceftamin's potential outliers: {outliers_Ceftamin}")
```

```python
# Generate a box plot that shows the distrubution of the tumor volume for each treatment group.
flierprops = dict(marker='o', markerfacecolor='r', markersize=12,linestyle='none', markeredgecolor='b')
plt.boxplot(final_tumour_data_by_treatment,flierprops=flierprops)
plt.ylabel("Final Tumour Volume (mm3)")
plt.xticks([1,2,3,4],treatments)
plt.plot()
```

## Line and Scatter Plots

```python
# Generate a line plot of tumor volume vs. time point for a single mouse treated with Capomulin

Capomulin_data = study_data_clean[study_data_clean["Drug Regimen"]=="Capomulin"]

random_mouse = random.randrange(1,len(Capomulin_data)-1)
mouse = Capomulin_data.iloc[random_mouse][0]              #get a random mouse"

Capomulin_data_mouse = Capomulin_data[Capomulin_data["Mouse ID"]==mouse] 

plt.plot(Capomulin_data_mouse["Timepoint"],Capomulin_data_mouse["Tumor Volume (mm3)"])
plt.ylabel("Tumour Volume (mm3)")
plt.xlabel("Timepoint (days)")
plt.title(f"Capomulin Treatment of mouse {mouse}")
plt.show()
```

```python
# Generate a scatter plot of mouse weight vs. the average observed tumor volume for the entire Capomulin regimen

mean_tumour_vol = Capomulin_data.groupby("Mouse ID")["Tumor Volume (mm3)"].mean()
mean_weight = Capomulin_data.groupby("Mouse ID")["Weight (g)"].mean()
plt.scatter(mean_weight,mean_tumour_vol)
plt.ylabel("Average Tumour Volume (mm3)")
plt.xlabel("Weight (g))")
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and a linear regression model 
# for mouse weight and average observed tumor volume for the entire Capomulin regimen

(slope,intercept,r_value,p_value,stderr) = st.linregress(mean_weight,mean_tumour_vol)
y_vals = slope * mean_weight + intercept

print(f"The correlation between mouse weight and the average tumor volume is {round(r_value,2)}" )

plt.scatter(mean_weight,mean_tumour_vol)
plt.plot(mean_weight, y_vals,color="red")
plt.ylabel("Average Tumour Volume (mm3)")
plt.xlabel("Weight (g))")
plt.show()
```
